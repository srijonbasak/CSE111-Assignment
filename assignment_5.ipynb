```python
#task 4
class Color:
    def __init__(self,clr):
        self.clr = clr
    def __add__(self,other):
        orng = ['red','yellow']
        vio = ['red','blue']
        green = ['yellow','blue']
        if (self.clr in orng )and (other.clr in orng) and (self.clr != other.clr):
            return Color('Orange')
        elif (self.clr in vio )and (other.clr in vio) and (self.clr != other.clr):
            return Color('Violet')
        elif (self.clr in green )and (other.clr in green) and (self.clr != other.clr):
            return Color('Green')
C1 = Color(input("First Color: ").lower())
C2 = Color(input("Second Color: ").lower())
C3 = C1 + C2
print("Color formed:", C3.clr)
```

```python
#task 5
class Circle:
    def __init__(self,r):
        self.__r = r
    def getRadius(self):
        return self.__r
    def area(self):
        return (22/7)*(self.__r**2)
    def __add__(self,other):
        return Circle(self.__r+other.__r)


c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" ,c1.area())
 
c2 = Circle(5)
print("Second circle radius:" ,c2.getRadius())
print("Second circle area:" ,c2.area())
 
c3 = c1 + c2 
print("Third circle radius:" ,c3.getRadius())
print("Third circle area:" ,c3.area())
```

```python
#task 6
class Triangle:
    def __init__(self,x,y):
        self.__x = x
        self.__y = y
    def getBase(self):
        return self.__x
    def getHeight(self):
        return self.__y
    def area(self):
        return self.__x*self.__y
    def __sub__(self,other):
        return Triangle(self.__x - other.__x, self.__y - other.__y)



t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())
 
t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
 
t3 = t1 - t2 
print("Third Triangle Base:" , t3.getBase())
print("Third Triangle Height:" , t3.getHeight())
print("Third Triangle area:" ,t3.area()) 
```

```python
#task 8
class Coordinates:
    def __init__(self,x,y):
        self.__x = x
        self.__y = y
    def __sub__(self,other):
        return Coordinates(self.__x - other.__x, self.__y - self.__y)
    
    def __mul__(self,other):
        return Coordinates(self.__x * other.__x, self.__y * self.__y)
    
    def __eq__(self,other):
        if (self.__x == other.__x and self.__y == self.__y):
            return ('The calculated coordinates are the same.')
        else:
            return ('The calculated coordinates are NOT the same.')
            
    def detail(self):
        return self.__x,self.__y
    



p1 = Coordinates(int(input()),int(input()))
p2 = Coordinates(int(input()),int(input()))

p4 = p1 - p2 
print(p4.detail())

p5 = p1 * p2
print(p5.detail())

point_check = (p4 == p5)
print(point_check)
```
