```python
target = input()
low_count = 0
up_count = 0
for letter in target:
    if ord(letter) in range(97,123):
        low_count += 1
    elif ord(letter) in range(65,91):
        up_count += 1
if low_count < up_count:
    target = target.upper()
else:
    target = target.lower()
print(target)
```

```python
target = input()
num_count = 0
wrd_count = 0
for letter in target:
    if ord(letter) in range(48,58):
        num_count += 1
    elif ord(letter) in range(65,91) or ord(letter) in range(97,123):
        wrd_count += 1
if num_count > 0 and wrd_count > 0:
    print("MIXED")
elif num_count > 0:
    print('NUMBER')
elif wrd_count > 0:
    print("WORD")
```

```python
password = input()
up_t = False
lw_t = False
nu_t = False
sp_t = False

for x in password:
    if ord(x) in range(65,91):
        up_t = True
    elif ord(x) in range(97,123):
        lw_t = True
    elif ord(x) in range(48,58):
        nu_t = True
    elif x in ['_','#','@','$']:
        sp_t = True
if up_t and lw_t and nu_t and sp_t:
    print('OK')
if not up_t:
    print('Uppercase Character missing', end = ', ')
if not lw_t:
    print('Lowercase Character missing', end = ', ')
if not nu_t:
    print('Digit missing', end = ', ')
if not sp_t:
    print('Special character missing', end = ', ')
```

```python
list1 = []
list2 = []

while True:
    num = input()
    if num.lower() == 'stop':
        break
    else:
        list1.append(num)
for ele in list1:
    if ele not in list2:
        print(f'{ele} - {list1.count(ele)} times')
        list2.append(ele)
```

```python
times = int(input())
result = []
for x in range(times):
    temp1 = []
    summ = 0
    temp = input().split()
    for num in temp:
        temp1.append(int(num))
        summ += int(num)
    result.append([temp1,summ])
lrg = result[0][1]
for nums in result:
    if nums[1] > lrg:
        lrg = nums[1]
print(lrg)
for num in result:
    if num[1] == lrg:
        print(num[0])
```
