```python
#task 1
class Student:
    idd = 1
    def __init__(self,name,dept,age,cg):
        self.__id = Student.idd
        Student.idd += 1
        self.__name = name
        self.__dept = dept
        self.__age = age
        self.__cg = cg
    def get_details(self):
        print(f'ID: {self.__id} \nName: {self.__name}\nDepartment: {self.__dept}\nAge: {self.__age}\nCGPA: {self.__cg}')
    @classmethod
    def from_String(cls,string):
        name, dept, age, cg = string.split('-')
        return Student(name,dept,age,cg)
        




s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details() 
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")

s4.get_details()
```

```python
#task 2
class Assassin:
    count = 1
    def __init__(self,name,rate):
        self.__count = Assassin.count
        Assassin.count += 1
        self.__name = name
        if rate == 100:
            self.__rate = '100%'
        else:
            self.__rate = f'{100-int(rate)}%'
    def printDetails(self):
        print(f'Name: {self.__name}\nSuccess rate: {self.__rate}\nTotal number of Assassin: {self.__count}')
    @classmethod
    def failureRate(cls,name,rate):
        return cls(name,rate)
    @classmethod
    def failurePercentage(cls,name,rate):
        rate = int(rate[:-1])
        return cls(name,rate)

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
#task 3
class Passenger:
    count = 0
    def __init__(self,name):
        Passenger.count += 1
        self.__name = name
        self.__weight = None
        self.__taka = 0
    def set_bag_weight(self,weight):
        self.__weight = weight
        if self.__weight < 21:
            self.__taka = 450
        elif self.__weight in range(21,51):
            self.__taka = 500
        elif self.__weight > 50:
            self.__taka = 550
    def printDetail(self):
        print(f'Name: {self.__name}\nBus Fare: {self.__taka} taka')
        


print('Total Passenger:', Passenger.count)
p1 = Passenger('Jack')
p1.set_bag_weight(90)
p2 = Passenger('Carol')
p2.set_bag_weight(10)
p3 = Passenger('Mike')
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print('Total Passenger:', Passenger.count)
```

```python
#task 5
class Employee:
    def __init__(self,name,exp):
        self.name = name
        self.workingPeriod = exp
    @classmethod
    def employeeByJoiningYear(cls,
                              name,yr):
        return cls(name,2022 - yr)
    @staticmethod
    def experienceCheck(yr, gen):
        if gen == 'male':
            gen = 'He'
        else:
            gen = 'She'
        if yr < 3:
            return f'{gen} is not experienced'
        else:
            return f'{gen} is experienced'


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#task 7
class Cat:
    Number_of_cats = 0
    def __init__(self,name,sec):
        self.name = name
        self.sec = sec
        Cat.Number_of_cats += 1
    def printCat(self):
        print(f'{self.name} cat is {self.sec}')
    def changeColor(self,name):
        self.name = name
    @classmethod
    def no_parameter(cls):
        return cls('White','sitting')
    @classmethod
    def first_parameter(cls,name):
        return cls(name,'sitting')
    @classmethod
    def second_parameter(cls,sec):
        return cls('White',sec)
    

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#task 8
class Cylinder:
    radius = 5
    height = 18
    def __init__(self,radius, height):
        print(f'Default radius={Cylinder.radius} and height={Cylinder.height}.')
        self.height = height
        self.radius = radius
        Cylinder.height = height
        Cylinder.radius= radius
        print(f'Updated: radius={Cylinder.radius} and height={Cylinder.height}.')
    @staticmethod
    def area(r,h):
        print(f'Area: {(2*(22/7)*(r**2))+(2*(22/7)*r*h)}')
        
    @staticmethod
    def volume(r,h):
        print(f'Volume: {(22/7)*(r**2)*h}')
        
    @classmethod
    def swap(cls,h,r):
        return cls(r,h)
    @classmethod
    def changeFormat(cls,string):
        r,h = string.split('-')
        return cls(int(r),int(h))
    

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
